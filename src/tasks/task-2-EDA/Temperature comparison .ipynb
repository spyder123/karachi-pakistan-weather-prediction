```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
```

```python
data = pd.read_csv('open_meteo.csv')
data
```

```python
data.describe()
```

```python
data = data[['time', 'city', 'temperature_2m_min', 'temperature_2m_max', 'temperature_2m_mean', 'precipitation_sum', 'precipitation_hours', 'et0_fao_evapotranspiration']]
data
```

```python
data['time'] = pd.to_datetime(data['time'], format='%Y/%m/%d')
```

```python
start_date = '5-1-2022'
end_date = '12-1-2022'
df = data[(data['time'] >= start_date) & (data['time'] <= end_date)]
```

```python
df
```

### Which city has low temperature and high temperature? 

```python
low_temp = df['temperature_2m_min'].min()
df[df['temperature_2m_min'] == low_temp]
```

We can see that Bhuj has low temperature than the other cities.
The city was not too hot at 2022 November

```python
high_temp = df['temperature_2m_max'].max()
df[df['temperature_2m_max'] == high_temp] 
```

From the above, The maximum temperature recorded in Bhuj was 42.2°C. The months of May and June are indeed known for higher temperatures and heatwaves in Bhuj.

### Minimum Temperature and Maximum Temperature Plot for 2022

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))

for city in df['city'].unique():
    data = df[df['city'] == city]
    ax1.plot(data['time'], data['temperature_2m_min'], label=city)
ax1.set_title('Minimum Temperature for each day')
ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature')
ax1.legend()

for city in df['city'].unique():
    data = df[df['city'] == city]
    ax2.plot(data['time'], data['temperature_2m_max'], label=city)
ax2.set_title('Maximum Temperature for each day')
ax2.set_xlabel('Time')
ax2.set_ylabel('Temperature')
ax2.legend()

for city in df['city'].unique():
    data = df[df['city'] == city]
    ax3.plot(data['time'], data['temperature_2m_mean'], label=city)
ax3.set_title('Average Temperature for each day')
ax3.set_xlabel('Time')
ax3.set_ylabel('Temperature')
ax3.legend()

plt.tight_layout()
plt.show()
```

From the above plot, we can see that all of the cities temperature. For instance, in 2020, in July, the minimum temperature is 28.9 degrees Celsius and the high temperature is 30 degrees Celsius in May. From these plots, we can conclude that the high temperature occurs around May, June, and October.

### Minimum Temperature and Maximum Temperature Plot for 2021

```python
start_date = '5-1-2021'
end_date = '12-1-2021'
df2 = data[(data['time'] >= start_date) & (data['time'] <= end_date)]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

for city in df2['city'].unique():
    data = df2[df2['city'] == city]
    ax1.plot(data['time'], data['temperature_2m_min'], label=city)
ax1.set_title('Minimum Temperature for each day')
ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature')
ax1.legend()

for city in df2['city'].unique():
    data = df2[df2['city'] == city]
    ax2.plot(data['time'], data['temperature_2m_max'], label=city)
ax2.set_title('Maximum Temperature for each day')
ax2.set_xlabel('Time')
ax2.set_ylabel('Temperature')
ax2.legend()

plt.tight_layout()
plt.show()
```

From the above graphs, we can clearly see that the high temperature in December 2021 was 30.1 degrees Celsius. From the previous graph, the high temperature in December 2022 was 28.7 degrees Celsius. That means last year's temperature was not as hot as 2021.

### Most heat waves period

```python
temperature_threshold = 30 
heatwave_periods = df[df['temperature_2m_max'] > temperature_threshold]
heatwave_periods
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for city in heatwave_periods['city'].unique():
    data = heatwave_periods[heatwave_periods['city'] == city]
    ax.plot(data['time'], data['temperature_2m_max'], label=city)

ax.set_title('Heat Wave Periods')
ax.set_xlabel('Time')
ax.set_ylabel('Maximum Temperature')
ax.legend()
plt.tight_layout()
plt.show()
```

Bhuj city causes heat waves other than those in other countries. Karachi doesn't have as many heatwaves as Bhuj. The high temperatures occur in May and June.

### Correlation for each city

```python
df.groupby('city').temperature_2m_min.describe()
```

Temperatures range from 18.5°C to 28.9°C, with an average around 24.5°C. The temperature distribution is somewhat narrow, with a standard deviation of 1.76°C.

```python
df.groupby('city').temperature_2m_max.describe()
```

The mean maximum temperature is about 31.5°C, with temperatures ranging from 27.7°C to 38.0°C. The distribution is relatively tight, with a standard deviation of 2.14°C.

```python
city_corr = pd.pivot_table(df, values='temperature_2m_max', index=['time'], columns=['city'])
city_corr
```

```python
city_corr.corr()
```

The correlation between Bhuj and Karachi has high correlation based on maximum temperature. When the maximum temperature in Bhuj tends to increase, the maximum temperature in Karachi also tends to increase, and vice versa.
