# Calculating precipitation sum for nearby cities

```python
start_date = '5-1-2021'
end_date = '11-1-2021'
filtered_df = data[(data['time'] >= start_date) & (data['time'] <= end_date)]

for city, group_data in filtered_df.groupby('city'):
    plt.plot(group_data['time'], group_data['precipitation_sum'], label=city)

plt.title('Precipitation Sum by City')
plt.xlabel('Date')
plt.ylabel('Precipitation')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Precipitation sum for karachi
